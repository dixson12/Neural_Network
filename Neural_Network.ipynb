```python
import pandas as pd
df=pd.read_csv('/content/breast-cancer.csv')
df
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['diagnosis']=le.fit_transform(df['diagnosis'])
```

```python
X=df.iloc[:,2:]
y=df.iloc[:,1]
y
```

```python
#Train_test_split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
#Scaling
from sklearn.preprocessing import MinMaxScaler
ms=MinMaxScaler()
ms.fit(X_train)
X_train=ms.transform(X_train)
X_test=ms.transform(X_test)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
ann=tf.keras.models.Sequential()
#input layer
ann.add(Dense(units=30,activation='relu'))
#Dense layer means it is fully connected layers. 
#Here in the brackets of Dense we give value for 'unit' as the number of features that we have in the data set

```

```python
#first hidden layer
ann.add(Dense(units=8,activation='relu'))
```

```python
#Add second hidden layer
ann.add(Dense(units=6,activation='relu'))
```

```python
#Add third hidden layer
ann.add(Dense(units=4,activation='relu'))
```

```python
#Output layer 
ann.add(Dense(units=1,activation='sigmoid'))
```

```python
ann.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
#Optimizer is used to update weight of features. Here we use adam optimizer
ann.fit(X_train,y_train,batch_size=32,epochs=100)
```

```python
ann.summary()
```

```python
y_pred=ann.predict(X _test)
```

```python
y_pred=y_pred>0.5
y_pred
from sklearn.metrics import accuracy_score
print(accuracy_score(y_pred,y_test))
```
